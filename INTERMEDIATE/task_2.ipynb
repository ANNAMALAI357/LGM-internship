IMPORTING NECESSARY LIBRARIES

```python
import numpy as np 

import pandas as pd 

import sklearn

import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.metrics import confusion_matrix 

from sklearn.tree import DecisionTreeClassifier 

from sklearn.metrics import accuracy_score 

from sklearn.datasets import load_iris

from matplotlib import pyplot as plt

from sklearn import datasets

from sklearn import tree

from sklearn.metrics import  confusion_matrix

import sklearn.metrics as metrics
```

LOADING THE DATA SET FROM SKLEARN AND PREPROCESSING THE DATA SET

```python
iris_data = load_iris() 

iris=pd.DataFrame(iris_data.data)

#priting features name of iris data 

print ("Features Name : ", iris_data.feature_names) 

#shape of datasets 

print ("Dataset Shape: ", iris.shape) 

#first five sample 

print ("Dataset: ",iris.head())  
```

DATA VISUALISATION ON THE DATA SET

```python
# HISTOGRAM
iris.hist(figsize=(12,10),bins=15)
plt.show()
```

```python
#HEAT MAP
plt.figure(figsize=(8,8))
p = sns.heatmap(iris.corr(),annot=True,cmap='RdYlGn')
```

SPLITTING THE DATA

```python
#priting samples and target 

X = iris.values[:, 0:4] 

Y = iris_data.target

print(X)

print(Y)
```

```python
# Splitting the dataset into train and test 

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
```

CREATING THE DECISION TREE MODEL

```python
# Decision tree classifier 

clf= DecisionTreeClassifier(random_state = 100)

#fitting the training data

clf.fit(X_train, y_train)
```

PREDICTION AND ACCURACY TESTING

```python
# prediction on random data

X=[[6.4,1.8 ,6.6 ,2.1]]

Y_pred=clf.predict(X)

print(Y_pred)

# prediction on X_test (testing data )

Y_pred=clf.predict(X_test)

print(Y_pred)
```

```python
#Accuray of the model 

print("Accuracy:",metrics.accuracy_score(y_test, Y_pred))

cm=np.array(confusion_matrix(y_test,Y_pred))

cm
```

PLOTTING THE DECISION TREE

```python
#Tree plotting 

tree.plot_tree(clf)
```

```python
#Decision making in decision tree

text_representation = tree.export_text(clf)

print(text_representation)
```
